State

```python
df.columns
```

```python
# Total number of orders from top 10 states

orders_states = df.groupby(['State'], as_index = False) ['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)

sns.set (rc = {'figure.figsize': (17,5)})
sns.barplot(data = orders_states, x = 'State', y = 'Orders')

#for bar in ax.containers:
 #   ax.bar_label(bars)
```

```python
#Total amount/ sales from top 10 states

sales_states = df.groupby(['State'], as_index = False) ['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)

sns.set(rc = {'figure.figsize' : (17,5)})

sns.barplot(data = sales_states, x = 'State', y = 'Amount' )
```

*From above graphs we can see that unexpectedly most of the orders are from Uttar Pradesh, Maharashtra, and Karnataka
respectively but total sales/amount is from UP, Karnataka, Maharashtra*

```python
ax = sns.countplot(data = df, x= 'Shaadi')

sns.set(rc = {'figure.figsize' : (5,3)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_states = df.groupby(['Shaadi', 'Gender'], as_index = False) ['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.set(rc = {'figure.figsize' : (7,5)})

sns.barplot(data = sales_states, x = 'Shaadi', y = 'Amount', hue=  'Gender')

#for bars in ax.containers:
 #   ax.bar_label(bars)
```

From above graphs we can see that that most of the buyers are married (women) and they have high purchasing power.

Ocuupation

```python
sns.set(rc= {'figure.figsize': (20,5)})

ax = sns.countplot(data = df, x= 'Occupation')

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
order_occupation = df.groupby(['Occupation'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False)

sns.set(rc = {'figure.figsize': (20,5)})
ax = sns.barplot(data = order_occupation, x = 'Occupation', y = 'Orders')

for bars in ax.containers:
    ax.bar_label(bars)
```

From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector.

Product Category

```python
sns.set(rc = {'figure.figsize' : (25,5)})

ax = sns.countplot(data = df, x = 'Product_Category')

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_prod = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)

sns.set(rc= {'figure.figsize': (20,5)})

ax = sns.barplot(data = sales_prod, x ='Product_Category', y = 'Amount')

#for bar in ax.containers:
 #   ax.bar_label(bar)
```

From above graphs we can see that most of the sold products are from Food, Clothing, Electrpnics

```python
#Top 10 most sold Product

sales_Pid = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)

sns.set(rc = {'figure.figsize' : (20,5)})

sns.barplot(data = sales_Pid, x= 'Product_ID', y = 'Orders')
```

Conclusion:
    
Married women age group 26-35 years from UP, Maharastra and karnataka working in IT, Healthcare and Aviation are more likely
to buy products from food, clothing and electronics category.
